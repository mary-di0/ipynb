```python
from sympy.abc import s

a = [1, 100, 500, 1000]
eps = 1e-5

def func(x1, x2, a):
    return x1**2 + a*x2**2

def f_grad(x1, x2, a):
    return np.array([2*x1, 2*a*x2])

s_0 = 0.1
x1, x2 = sp.symbols('x1 x2')
variables = [x1, x2]  
initial_point = {x1: 0, x2: 5}

def newton_step_optimization(phi, s_guess, eps=1e-5, max_iter=100):
    """Упрощенный метод Ньютона для поиска оптимального шага"""
    phi_prime = sp.diff(phi, s)
    phi_prime_prime = sp.diff(phi_prime, s)
    
    f_prime = sp.lambdify(s, phi_prime, "numpy")
    f_prime_prime = sp.lambdify(s, phi_prime_prime, "numpy")
    
    s_current = s_guess
    for _ in range(max_iter):
        f_p = f_prime(s_current)
        f_pp = f_prime_prime(s_current)
        
        if abs(f_p) < eps:
            break
            
        s_new = s_current - f_p / f_pp
        if abs(s_new - s_current) < eps:
            break
            
        s_current = s_new
    
    return float(s_current)

def fastest_descent(function, variables, initial_point, tolerance=0.1, max_iter=100):
    gradient = [sp.diff(function, var) for var in variables]
    gradient_funcs = [sp.lambdify(variables, grad, "numpy") for grad in gradient]
    objective_func = sp.lambdify(variables, function, "numpy")  
    
    x_current = np.array([initial_point[var] for var in variables])
    x_list = [x_current]
    f_list = [objective_func(*x_current)]   
    
    for i in range(max_iter):
        grad_values = np.array([grad(*x_current) for grad in gradient_funcs])
        print(f"Итерация {i+1}:")
        print(f"Точка: {x_current}, Значение функции: {f_list[-1]:.4f}")
        print(f"Градиент: {grad_values}")
        
        grad_norm = np.linalg.norm(grad_values)
        if grad_norm < tolerance:
            print(f"norm(grad) = {grad_norm:.4f}")
            print("Условие остановки выполнено. Минимум найден.")
            break 
            
        d_k = -grad_values / grad_norm
        
        # Создаем функцию для оптимизации шага
        phi = function.subs({
            var: (x_current[i] + s * d_k[i]) 
            for i, var in enumerate(variables)
        }).simplify()
        
        # Используем упрощенный метод Ньютона
        learning_rate = newton_step_optimization(phi, s_0, eps)
        
        x_new = x_current + learning_rate * d_k
        f_new = objective_func(*x_new)  
        
        print(f"Направление d_k: {d_k}")
        print(f"Оптимальный шаг s_k: {learning_rate:.6f}")
        print(f"Новая точка: {x_new}, Новое значение функции: {f_new:.4f}")
        print("-" * 50)     
        
        x_current = x_new
        x_list.append(x_current)
        f_list.append(f_new)        
    
    return x_list, f_list

# Пример вызова
result = fastest_descent(func(x1, x2, a[3]), variables, initial_point, eps, max_iter=1000)
```

```python
a = [1, 100, 500, 1000]
eps = 1e-5

def func(x1, x2, a):
    return x1**2 + a*x2**2

def f_grad(x1, x2, a):
    return np.array([2*x1, 2*a*x2])

s_0 = 0.1
x1, x2 = sp.symbols('x1 x2')
variables = [x1, x2]  
initial_point = {x1: 0, x2: 5}

def conjugate_gradient(function, variables, initial_point, tolerance=1e-5, max_iter=100):
    gradient = compute_gradient(function, variables)
    gradient_funcs = [sp.lambdify(variables, grad, "numpy") for grad in gradient]
    objective_func = sp.lambdify(variables, function, "numpy")  
    x_current = np.array([initial_point[var] for var in variables])
    x_list = [x_current]
    f_list = [objective_func(*x_current)]   
    p_k = None
    
    for i in range(max_iter):
        grad_values = np.array([grad(*x_current) for grad in gradient_funcs])
        condition = np.linalg.norm(grad_values)
        print(f"\nИтерация {i+1}:")
        print(f"Точка: {x_current}, Значение функции: {f_list[-1]:.10f}")
        print(f"Градиент: {grad_values}, Норма: {condition:.10f}")
        
        if condition < tolerance:
            print(f"\nУсловие остановки: норма градиента = {condition:.10f} < {tolerance}")
            print("Минимум найден.")
            break
            
        if i == 0:
            p_k = -grad_values
        else:
            # Формула Флетчера-Ривса
            t = np.dot(grad_values, grad_values) / np.dot(grad_prev, grad_prev)
            p_k = -grad_values + t * p_k
            
        # Оптимизация шага методом Ньютона
        new_coords = [x_current[j] + s * p_k[j] for j in range(len(variables))]
        step_function = function.subs({var: new_coords[j] for j, var in enumerate(variables)})
        step_function = step_function.simplify()
        
        learning_rate = newton_step_optimization(step_function, s_guess=0.1, eps=eps)
        
        x_new = x_current + learning_rate * p_k
        grad_prev = grad_values.copy()  # Сохраняем предыдущий градиент
        f_new = objective_func(*x_new)
        
        print(f"Направление p_k: {p_k}")
        print(f"Оптимальный шаг s_k: {learning_rate:.10f}")
        print(f"Новая точка: {x_new}, Новое значение функции: {f_new:.10f}")
        print(f"Изменение функции: {f_new - f_list[-1]:.10f}")
        
        x_current = x_new
        x_list.append(x_current)
        f_list.append(f_new)
    
    return x_list, f_list

conjugate_gradient(func(x1, x2, a[3]), variables, initial_point, eps, max_iter=100)
```
