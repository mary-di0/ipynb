```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(10, 60, num=6)
y = [1.06, 1.33, 1.52, 1.68, 1.81, 1.91]

plt.scatter(x, y)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid(True)
plt.show()
```

```python
f = np.array([[1, 10, 100], [1, 20, 400], [1, 30, 900], [1, 40, 1600], [1, 50, 2500], [1, 60, 3600]])
f_t = f.T
inv_ftf = np.linalg.inv(f_t.dot(f))
Y = np.array([[1.06], [1.33], [1.52], [1.68], [1.81], [1.91]])
fty = f_t.dot(Y)
result = inv_ftf.dot(fty)
print(result)
```

```python
def f(x):
    return result[2]*x**2+result[1]*x+result[0]

plt.scatter(x, y)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid(True)

plt.plot(x, f(x), color='green')
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid(True)

coeff = [0.785, 0.0303393, -0.000194643]
plt.plot(x, coeff[2]*x**2+coeff[1]*x+coeff[0], color='black')

plt.show()
```

```python
import scipy as sp
def mapping(x, betta_0, betta_1, betta_2):
    return betta_2*x**2 + betta_1*x + betta_0
popt, _ = sp.optimize.curve_fit(mapping, x, y)
popt
```

```python
plt.plot(x, mapping(x, *popt), color='red')

plt.scatter(x, y)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid(True)

plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid(True)

plt.plot(x, f(x), color='green', label='Numpy')
plt.plot(x, coeff[2]*x**2+coeff[1]*x+coeff[0], color='black', label='Analytical')
plt.plot(x, mapping(x, *popt), color='red', label='Scipy')

plt.legend(loc='best', fontsize=12)

plt.show()
```
