```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем 3D-график
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
#ax = plt.axes(projection='3d')

# 1. Подготовка данных для функции
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = (X - Y)**2 + 10*(X + 5)**2  # Функция f(x,y)

# 2. Визуализация функции 
ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.8)

# 3. Ограничение a) x + y >= 0 (полупространство)
x_plane = np.linspace(-10, 10, 10)
y_plane = np.linspace(-10, 10, 10)
X_plane, Y_plane = np.meshgrid(x_plane, y_plane)
Z_plane = np.zeros_like(X_plane)

# Определяем область, где x + y >= 0
mask = X_plane + Y_plane >= 0
X_plane, Y_plane, Z_plane = X_plane[mask], Y_plane[mask], Z_plane[mask]

# Добавляем полупространство на график
ax.plot_trisurf(X_plane, Y_plane, Z_plane, color='cyan', alpha=0.2)


# 4. Ограничение б) x = 1 - y (вертикальная плоскость)
x = np.linspace(-10, 10, 100)
z = np.linspace(0, 2000, 1000)
X, Z = np.meshgrid(x, z)
Y = 1 - X
ax.plot_surface(X, Y, Z, alpha=0.3, color='red')


# Настройки графика
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('f(x,y)', fontsize=12)


# Улучшенный угол обзора
ax.view_init(elev=30, azim=45)


plt.show()
```

```python
import scipy as sp
import pandas as pd
def func(x, y):
    return (x-y)**2 + 10*(x+5)**2

def h(x, y):
    return x+y-1

def g(x, y):
    return -x-y

def phi(x, y, pk):
    return pk/2*(h(x, y)**2 + (max(g(x, y), 0)**2))

def fk(point, pk):
    x, y = point
    return func(x, y) + phi(x, y, pk)

parameters = [1, 2, 5, 100, 1000]
epses = [0.5, 0.1, 0.01, 0.0001, 0.00001]
c = 2
results = []

def method(x, y, eps, p0):
    k = 1
    pk = p0*c
    res = np.array([x, y])
    while phi(res[0], res[1], pk) > eps:
        res = sp.optimize.minimize(lambda x: fk(x, pk), res).x
        pk *= c
        k += 1
    x_sol, y_sol = res[0], res[1]
    return (x_sol, y_sol), func(x_sol, y_sol)

for eps in epses:
    result_line = []
    for p0 in parameters:
        solution, f_val = method(-1, 0, eps, p0)
        result_line.append((solution, f_val))
    results.append(result_line)

coord_data = []
func_data = []

for i, eps in enumerate(epses):
    coord_row = []
    func_row = []
    for j in range(len(parameters)):
        coord_row.append(f"({results[i][j][0][0]:.6f}, {results[i][j][0][1]:.6f})")
        func_row.append(f"{results[i][j][1]:.6f}")
    coord_data.append(coord_row)
    func_data.append(func_row)


df_coord = pd.DataFrame(
    coord_data,
    index=[f"ε={eps}" for eps in epses],
    columns=[f"p₀={p}" for p in parameters]
)
df_coord.index.name = "Точность"
df_coord.columns.name = "Коэффициент штрафа"


df_func = pd.DataFrame(
    func_data,
    index=[f"ε={eps}" for eps in epses],
    columns=[f"p₀={p}" for p in parameters]
)
df_func.index.name = "Точность"
df_func.columns.name = "Коэффициент штрафа"


print("Координаты решений:")
display(df_coord)

print("\nЗначения функции f(x,y):")
display(df_func)
```

```python
def func(x, y):
    return (x-y)**2 + 10*(x+5)**2

def g(x, y):
    return -x-y

def phi(x, y, pk):
    if(g(x, y) < 0):
        return -pk/g(x,y)
    else:
        return 100000

def fk(point, pk):
    x, y = point
    return func(x, y) + phi(x, y, pk)

parameters = [0.5, 1, 2, 5, 100]
epses = [0.5, 0.1, 0.01, 0.0001, 0.00001]
c = 2
results = []

def method(x, y, eps, p0):
    k = 1
    pk = p0
    res = np.array([x, y])
    while phi(res[0], res[1], pk) > eps:
        res = sp.optimize.minimize(lambda x: fk(x, pk), res).x
        pk /= c
        k += 1
    x_sol, y_sol = res[0], res[1]
    return (x_sol, y_sol), func(x_sol, y_sol)

for eps in epses:
    result_line = []
    for p0 in parameters:
        solution, f_val = method(-2, 3, eps, p0)
        result_line.append((solution, f_val))
    results.append(result_line)

coord_data = []
func_data = []

for i, eps in enumerate(epses):
    coord_row = []
    func_row = []
    for j in range(len(parameters)):
        coord_row.append(f"({results[i][j][0][0]:.6f}, {results[i][j][0][1]:.6f})")
        func_row.append(f"{results[i][j][1]:.6f}")
    coord_data.append(coord_row)
    func_data.append(func_row)


df_coord = pd.DataFrame(
    coord_data,
    index=[f"ε={eps}" for eps in epses],
    columns=[f"p₀={p}" for p in parameters]
)
df_coord.index.name = "Точность"
df_coord.columns.name = "Коэффициент pk"


df_func = pd.DataFrame(
    func_data,
    index=[f"ε={eps}" for eps in epses],
    columns=[f"p₀={p}" for p in parameters]
)
df_func.index.name = "Точность"
df_func.columns.name = "Коэффициент pk"


print("Координаты решений:")
display(df_coord)

print("\nЗначения функции f(x,y):")
display(df_func)
```
